```python
import xml.etree.ElementTree as ET
xtree = ET.parse("user_data.xml")
xroot=xtree.getroot()
cols = ["firstName", "lastName", "age", "sex", "retired", "dependants", "marital_status", "salary", "pension", "company", "commute_distance", "address_postcode"]
rows = []
for node in xroot:
    s_firstName=node.attrib.get("firstName")
    s_lastName=node.attrib.get("lastName")
    s_age=node.attrib.get("age")
    s_sex=node.attrib.get("sex")
    s_retired=node.attrib.get("retired")
    s_dependants=node.attrib.get("dependants")
    s_marital_status=node.attrib.get("marital_status")
    s_salary=node.attrib.get("salary")
    s_pension=node.attrib.get("pension")
    s_company=node.attrib.get("company")
    s_commute_distance=node.attrib.get("commute_distance")
    s_address_postcode=node.attrib.get("address_postcode")
    rows.append({"firstName": s_firstName, "lastName": s_lastName, "age": s_age, "sex": s_sex,"retired": s_retired, "dependants": s_dependants, "marital_status": s_marital_status, "salary": s_salary, "pension": s_pension, "company": s_company,"commute_distance": s_commute_distance, "address_postcode": s_address_postcode})

data3 = pd.DataFrame(rows, columns = cols)
```
